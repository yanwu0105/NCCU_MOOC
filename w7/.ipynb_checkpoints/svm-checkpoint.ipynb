```python
%matplotlib inline 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# 這是Kaggle上，預測「鐵達尼號存活率」的，非常經典的例子，
# 我們今天的目標，是利用其中的「性別 Sex」跟「年齡 Age」來預測「存活與否 Survived」(1是存活，0則否）。

df = pd.read_csv("train.csv") 
df.head(5)
```

```python
# 首先，請取出所有的Age跟Sex作為題目，和Survived做為答案並分成三個同名的numpy array。  (10分)
# (提示： Age.shape = (891, ))
Age = np.array(df.Age)
Sex = np.array(df.Sex)
Survived = np.array(df.Survived)
Age.shape = (-1,)
Sex.shape = (-1,)
Survived.shape = (-1,)
```

```python
# 其中，由於Age有一些值是空值NaN，請把這些空值都先換成0。   (10分)
# （提示： 可使用 np.isnan(某個你想確認是不是NaN的年齡) ）
Age = np.nan_to_num(Age) 
```

```python
type(Sex[1])
```

```python
Sex[0] == "male"
```

```python
# 再來，Sex部分因為都是文字，不方便我們做svm，請把Male都換成0，Female都換成1   （10分）
Sex = list(Sex)
Sex1 = []

for i in Sex:
    if i == "male" :
        Sex1.append(0)
    elif i == "female":
        Sex1.append(1)
    Sex = np.array(Sex1)
```

```python
Sex
```

```python
# 由於我們想把Sex跟Age都當成data，請把Sex和Age兩個array合併成一個。（10分）
# （提示：先用np.reshape，再用np.append）
Sex = np.reshape(Sex,(-1,1))
Age = np.reshape(Age,(-1,1))
X = np.concatenate((Sex,Age), axis=1)
Y = Survived
```

```python
# 實際拿svm 訓練看看，並用同樣的data，preict看看結果（這次不用分training data跟testing data）
#（這次是因為前面做過了，一般都還是要噢！）(10分)
from sklearn.svm import SVC
```

```python
clf = SVC()
clf.fit(X,Y)
```

```python
a = list(clf.predict(X) - Y)
```
