# 第三題_錄取分數表格

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
score =  [86.5, 79.4, 88.2, 96.5, 86.3, 76.8, 96.1, 78.5, 82.4, 88.6]
collage = {
"土語系": [2, 2, 0, 1, 0, 0, 0, 1, 1, 0],
"財政系": [1, 1.5, 0, 1.5, 0, 0, 0, 0, 0, 0],
"金融系": [1, 1.5, 0, 2, 0, 0, 0, 0, 0, 0],
"國貿系": [1.5, 2, 0, 2, 0, 0, 0, 0, 0, 0],
"會計系": [1.5, 2, 0, 2, 0, 0, 0, 1, 1, 0],
"法律系": [1, 2, 0, 1, 0, 0, 0, 1, 0, 1],
"公行系": [1, 1.5, 0, 1.25, 0, 0, 0, 0, 0, 1.25],
"應數系": [1, 1, 1.5, 0, 1, 1, 0, 0, 0, 0],
"教育系": [1.5, 2, 0, 1.25, 0, 0, 0, 1, 1, 0]
} 
marjor= ["國文", "英文", "數甲", "數乙", "物理", "化學", 
         "生物", "歷史", "地理", "公民"]
```

```python
col = list()
for x in collage:
    a = collage[x]
    col.append(a)
col_name = list(collage.keys())
```

```python
df = pd.DataFrame(col,index=col_name,columns=marjor)
```

```python
df
```

```python
a = df * score
a["加權分數"] = a[marjor].sum(1)
a
```

```python
df1 = df
df1["加權分數"] = a["加權分數"]
df1
```

# 迴圈跑法1

```python
total = list()
for x in col_name:
    s = np.array(score)
    c = np.array(collage[x])
    total.append(np.dot(s,c)) 
total
```

```python
df2 = df
df2["加權分數"] = total
df2
```

# 迴圈跑法2

```python
df3 = df
for x in col_name:
    s = np.array(score)
    c = np.array(collage[x])
    df3.loc[x,"加權分數"] = np.dot(s,c)
df3
```
