### 篩條件

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = np.array([3,-2,4,-5,333,-555,-60])
```

```python
L<0
```

```python
L[L<0]
#直接篩條件
```

```python
x = np.linspace(-5,5,500)
y = np.sinc(x)
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.plot(x[y>0],y[y>0],"o")
```
